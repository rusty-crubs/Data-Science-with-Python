# Sallen Key (S-K Biquad Circuit)

```python
import matplotlib.pyplot as plt
import numpy as np
# from control import tf, bode
import control as ctrl
```

```python
r1= 0.707
r2 = 1.414
c1 = 1
c2 = 1
ra = 4.2426
rb = 4.2426
k=1
num = [0, 0, 0, k/(r1*r2*c1*c2)]
den = [0, 1, (1/(r1*c1) + 1/(r2*c2) + 1/(r1*c1) - k/(r2*c2)), 1/(r1*c1*r2*c2)]
```

```python
# System
system = ctrl.TransferFunction(num, den)
system
```

```python
# setting w
w = np.logspace(-1,6,25000)
w
```

```python
# bode
mag, phase, omega = ctrl.bode(system, omega=w, plot=False)

# Plot
plt.plot(omega,np.log10(mag))
```
