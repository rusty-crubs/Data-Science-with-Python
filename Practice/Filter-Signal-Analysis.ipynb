# Transfer Functions Plots
Filter Design for various Transfer Function

```python
import matplotlib.pyplot as plt
import numpy as np
from control import tf, bode
```

```python
# Creating Transfer Funtion Numerator and Denominator
Numerator = [2, 0, 0, 0]
Denominator = [ 1, 6, 6, 2]
# Creating Transfer Function
System = tf(Numerator, Denominator)

# Creating frequency range
w = np.logspace(-1,6,2000) # From range 0.1 to 1e-6

# Bode the Magnitude and Phase
magnitude, phase, omega = bode(System, omega=w, plot=False)

# Plot
plt.figure(figsize = (10, 7))

# Magnitude
plt.subplot(1,1,1)
plt.semilogx(omega, 20 * np.log10(np.squeeze(magnitude)), 'b', linewidth=2)
plt.grid()
plt.ylabel('Magnitude')
plt.title('Question-1')
plt.show()
```

# Tow Thomos Biquad

## Low Pass Filter

```python
# Component Values of Low Pass Circuit
R1 = 5
R2 = 1
R3 = 0.667
R4 = 1
C1 = 1e-4
C2 = 1e-4
# Transfer Function numerator and denominator
Numerator = [0, 0, 0, -1/(R3*R4*C1*C2)]
Denominator = [0, 1, 1/(C1*R1), 1/(R2*R4*C1*C2)]

# Creating Transfer Function
System = tf(Numerator, Denominator)

# Creating Frequency Vector
w = np.logspace(-1,6,2000) # From 0.1 to 1e6 rad/s

# Bode magnitude, and phase
mag, phase, omega = bode(System, omega = w, plot=False)

# Plot (Very close to MatLab)
plt.figure(figsize=(10,7))

# Magnitude plot
# plt.subplot(2,1,1)
plt.semilogx(omega, 20 * np.log10(np.squeeze(mag)),'b',linewidth=2)
plt.grid()
plt.ylabel("Magnitude (dB)")
plt.title("Low Pass Filter")

plt.show()
```

# High Pass Filter

```python
# High Pass Filter components value
R1 = 5
R2 = 1
R4 = 1
C1 = 1e-4
C2 = 1e-4

# Creating Transfer Funtion Numerator and Denominator
Numerator = [0,1,0,0]
Denominator = [0, 1, 1 / ( R1 * C1 ), 1 / ( R2 * R4 * C1 * C2)]

# Creating Transfer Function
System = tf(Numerator, Denominator)

# Creating a frequency range
w = np.logspace(-1,6,2000) # From 0.1 to 1e6 rad/s

# Bode magnitude and phase
mag, phase, omega = bode(System, omega=w, plot=False)

# Plot 
plt.figure(figsize=(10,7))

# Magnitude
plt.semilogx(omega, 20 * np.log10(np.squeeze(mag)), 'b', linewidth=2)
plt.grid()
plt.ylabel("Magnitude[dB]")
plt.title("High Pass Filter")
plt.show()
```

## Band Pass Filter

```python
# Band Pass components value
R1 = 5
R3 = 3.33
R2 = 1
R4 = 1
C1 = 1e-4
C2 = 1e-4

# Creating Transfer Function Numerator and Denominator
Numerator = [0, 0, -1/(R3*C1), 0]
Denominator = [0, 1, 1/(R1*C1), 1/(R2*R4*C1*C2)]

# Create a Frequency
w = np.logspace(-1, 6, 2000) # From 0.1 to 1e6 rad/s

# Creating Transfer Function
System = tf(Numerator, Denominator)

# Bode magnitude and phase
mag, phase, omega = bode(System, omega = w, plot=False)

# Plot
plt.figure(figsize=(10,7))

# Magnitude
plt.semilogx(omega, 20 * np.log10(np.squeeze(mag)), 'b', linewidth=2)
plt.grid()
plt.ylabel('Magnitude[dB]')
plt.title("Band Pass Filter")
plt.show()
```

## Band Stop Filter

```python
# Band Pass components value
R1 = 5
R2 = 1
R4 = 1
C1 = 1e-4
C2 = 1e-4

# Creating Transfer Function Numerator and Denominator
Numerator = [0, 1, 0, 1/(R2*R4*C1*C2)]
Denominator = [0, 1, 1/(R1*C1), 1/(R2*R4*C1*C2)]

# Create a Frequency
w = np.logspace(-1, 6, 2000) # From 0.1 to 1e6 rad/s

# Creating Transfer Function
System = tf(Numerator, Denominator)

# Bode magnitude and phase
mag, phase, omega = bode(System, omega = w, plot=False)

# Plot
plt.figure(figsize=(10,7))

# Magnitude
plt.semilogx(omega, 20 * np.log10(np.squeeze(mag)), 'b', linewidth=2)
plt.grid()
plt.ylabel('Magnitude[dB]')
plt.title("Band Stop Filter")
plt.show()
```

## All Pass Filter

```python
# --- Component Values ---
# Using your calculated values for f0 = 1kHz, Q = 0.707
R1 = 11.2e3
R2 = 15.9e3
R4 = 15.9e3
C1 = 1e-8
C2 = 1e-8

# --- Calculating Coefficients ---
# Note: In control.tf, [a, b, c] represents as^2 + bs + c. 
# We remove the leading 0 to avoid "singular system" errors in some libraries.
num_coeffs = [1, -1/(R1*C1), 1/(R2*R4*C1*C2)]
den_coeffs = [1, 1/(R1*C1), 1/(R2*R4*C1*C2)]

# --- Creating Transfer Function ---
System = tf(num_coeffs, den_coeffs)

# --- Frequency Range ---
# 10Hz to 100kHz is a better range for a 1kHz center frequency
w = np.logspace(1, 6, 2000) 

# --- Simulation ---
mag, phase, omega = bode(System, omega=w, plot=False)

# --- Enhanced Plotting ---
# fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
plt.figure(figsize=(10, 8))

# Magnitude Plot (Should be a flat line at 0dB)
# ax1 = plt.gca()
# ax1.semilogx(omega, 20 * np.log10(mag), color='blue', linewidth=2)
# ax1.set_ylabel('Magnitude (dB)')
# ax1.set_title('Bode Plot: Second-Order All-Pass Filter')
# ax1.grid(True, which="both", ls="-", alpha=0.5)
# ax1.set_ylim(-1, 1) # Zoom in to show it is flat

plt.semilogx(omega, 20 * np.log10(mag), color='blue', linewidth=2)
plt.ylabel('Magnitude (dB)')
plt.title('Bode Plot: Second-Order All-Pass Filter')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.ylim(-1, 1) # Zoom in to show it is flat

# Phase Plot (The key characteristic)
# Converting phase from radians to degrees
# ax2.semilogx(omega, np.degrees(phase), color='red', linewidth=2)
# ax2.set_ylabel('Phase (degrees)')
# ax2.set_xlabel('Frequency (rad/s)')
# ax2.grid(True, which="both", ls="-", alpha=0.5)

# plt.tight_layout()
plt.show()
```

# Sallen Key (S-K Biquad Circuit)

```python
r1= 0.707
r2 = 1.414
c1 = 1
c2 = 1
ra = 4.2426
rb = 4.2426
k=1
```

```python
def system():
    num = [0, 0, 0, k/(r1*r2*c1*c2)]
    den = [0, 1, (1/(r1*c1) + (1-k)/(r2*c2) + 1/(r2*c1)), 1/(r1*c1*r2*c2)]
    System = tf(num, den)
    return System
```

```python
# System
sys = system()
print(sys)
```

```python
# Creating frequency range
w = np.logspace(-1,6,2000) # From range 0.1 to 1e-6
print(w)
```

```python

# Bode the Magnitude and Phase
magnitude, phase, omega = bode(sys, omega=w, plot=False)

# Magnitude
fig, (mag_plot,phase_plot) = plt.subplots(2, 1, figsize=(10,8), sharex = True)
mag_plot.semilogx(omega, 20 * np.log10(magnitude), 'b', linewidth=2)
mag_plot.set_ylabel('Magnitude')
mag_plot.set_title('Sallen-Key Filter')
# plt.ylim(-1,1)
mag_plot.grid(True)

phase_plot.semilogx(omega, np.degrees(phase), 'r', linewidth=2)
phase_plot.set_ylabel('Phase (degrees)')
phase_plot.set_xlabel('Frequency (rad/s)')
phase_plot.grid(True)
plt.tight_layout()
plt.show()
```

# Frequency Transform
Converting a 1st order butterworth LPF to HPF, BPF, BSF

```python
num = [0, 0, 0, 1]
den = [0, 0, 1, 4]
System = tf(num, den)
w = np.logspace(-1,6,2000) # From range 0.1 to 1e-6
mag, phase, omega = bode(System, omega=w, plot=False)
plt.figure(figsize=(8,4))
plt.semilogx(omega, 20 * np.log10(mag), color='blue', linewidth=2)
plt.title('Conversion of LPF to LPF')
plt.ylabel('Magnitude (dB)')
plt.grid()
```
