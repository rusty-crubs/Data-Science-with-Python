# Chapter 1: Introduction to Python
Python is one of the most popular programming languages. It's simple to use, packed with features and support by wide range of libraries and frameworks. Its clean syntax makes it beginner-friendly.

- A high-level language, used in data science, automation, AI, Web development and more.
- Known for its readibility, which means code is easier to write, understand and maintain.
- Backed by strong library support, we don't have to build everything from scratch (like C/C++).
