- A labeled array (like a column in Excel).

### 1. **Read CSV**
- CSV files are the Comma Seperated Files that allow storage of tabular data.
- To access data from the CSV fiel, you require **read_csv()** from Pandas that retrives data in the form of the data frame.

```python
peoples = 'https://media.geeksforgeeks.org/wp-content/uploads/20241121154629307916/people_data.csv'
```

- Pandas allows you to directly read a CSV file hosted on the internet using the file url.
- The following URL (**people_data.csv** from geeksforgeeks) is indicated by **peoples**

```python
import pandas as pd
df = pd.read_csv(peoples)
df
```

- **read_csv()** function in Pandas is used to read data from CSV file into a Panda DataFrame.
- CSV files are plain-text files where each row represents a record, and columns are separated by commas.
- **Syntax**:
> pd.read_csv(filepath_or_buffer, sep=',', header='infer', index_col=None, usecols=None, engine=None, skiprows=None, nrows=None)

- **Parameters**:
    - **filepath_or_buffer**: Location of the csv file. It accepts any string path or URL of the file.
    - **sep**: It stands for seperator, default is ',' (comma).
    - **header**: It accepts int, a list of int, row numbers to use as the column names, and the start of the data. If no names are passed, i.e., header=None, then, it will display the first column as 0, the second as 1 and so on.
    - **usecols**: Retrieves only selected columns from the CSV file.
    - **nrows**: Number of rows to be displayed from the dataset.
    - **index_col**: If None, there are no index numbers displayed along with records.
    - **skiprows**: Skips passed rows in the new data frame.

### Features in Pandas read_csv
1. Read specific columns using read_csv
    - The usecols parameter allows to load only specific columns from a CSV file. This reduces memory usage and processing time by importing only the required data.

```python
df = pd.read_csv(peoples, usecols=['First Name', 'Email'])
df
```

2. Setting an index Column (index_col)
    - The index_col parameter sets one or more columns as the DataFrame index, making the specified column(s) act as row labels for easier data referencing.

```python
df = pd.read_csv(peoples, index_col='First Name')
df
```

3. Handling Missing Values Using read_csv
    - The na_values parameter replaces specified strings (e.g "N/A", "Unknown") with NaN, enabling consistent handling of missing or incomplete data during analysis.

```python
df = pd.read_csv(peoples, na_values=["N/A", "Unknown"])
df
```

4. Reading CSV Files with Different Delimiters
    - In this example, we will take a CSV file and then add some special characters to see how the sep parameter works.

```python
import pandas as pd

data = """totalbill_tip, sex:smoker, day_time, size
16.99, 1.01:Female|No, Sun, Dinner, 2
10.34, 1.66, Male, No|Sun:Dinner, 3
21.01:3.5_Male, No:Sun, Dinner, 3
23.68, 3.31, Male|No, Sun_Dinner, 2
24.59:3.61, Female_No, Sun, Dinner, 4
25.29, 4.71|Male, No:Sun, Dinner, 4"""

with open("sample.csv", "w") as file:
    file.write(data)
print(data)
```

- The sample data is stored in a multi-line string for demonstration purpose.
    1. **Seperator** (sep): The sep='[:,|_]' argument allows Pandas to handle multiple delimiters (:,|,_,,) using a regular expression.
    2. **Engine**: The engine='python' argument is used because the default C engine does not support regular expressions for delimiters.

```python
df = pd.read_csv('sample.csv',
                 sep='[:, |_]',
                 engine='python')
df
```

## 5. Data Indexing and Selection
The most common mistake new data scientists make is misunderstanding how to slice data. Pandas provides two highly optimized access methods:

1. **.loc** (**Local-based**): You access rows and columns by their explicit labels (names).
2. **.iloc** (**Integer-based**): You access rows and columns by their implicit integer position (like a standard array).

```python
import numpy as np
data = {
    'Sensor_ID':['ESP_01', 'ESP_02', 'ESP_01', 'ESP_03', 'ESP_02'],
    'Temperature_c':[22.4, 23.4, 22.6, np.nan,23.0],
    'Status': ['active','active','active','offline','active']
}
df = pd.DataFrame(data)
```

```python
# Slicing and Dicing

# 1. Accessing a column ( returns a Series)
temperature = df['Temperature_c']

# 2. Accessing a row by index position using .iloc
first_row = df.iloc[0]
print(f"iloc:\n{first_row}")

# 3. Accessing a subset of rows and specific columns using .loc
# Format: df.loc[row_labels, column_labels]
active_sensors = df.loc[0:2,['Sensor_ID', 'Status']]
print(f"loc:\n{active_sensors}")
```

## Boolean Indexing (Filtering)
- To filter data, we pass a boolean array (a mask) directly into the DataFrame. This is highly optimized and avoids slow Python **for** loops.

```python
# Create a mask: Which rows have temperature greater than 22.5
high_temp_mask = df['Temperature_c'] > 22.5
print(f"Temperature greater than 22.5: {high_temp_mask}")
```

```python
# Apply the mask to filter the DataFrame
hot_sensors_df = df[high_temp_mask]
print(f"Filtered DataFrame: {hot_sensors_df}")
```

```python
# Pro-tip: Combine conditions using bitwise operations (& for AND, | for OR)
# Always wrap conditions in parantheses !!
critical_sensors = df[(df['Temperature_c']>22.5) & (df['Status'] == 'active')]
print(f'Critical value:\n {critical_sensors}')
```

## 6. Data Cleaning and Imputation
- Real-world data is messy. Sensors fail, network packets drop, and user inputs are flawed.
- Handling missing data (represented as NaN - Not a Number) is a critical step before feeding data into any machine learning model.
- Imputation is the fancy data science term for "filling in the blanks".

- The **isna()** function scans your DataFrame or Series and returns a "mask" of the same shape, where each cell is either True (missing) or False (not missing).
- In pandas, *isnull()* is just an alias for *isna()*.

```python
# Handling missing data
missing_mask = df.isna() # Returns a DataFrame of True/False
print(f"Missing masks: {missing_mask}")
```

- Counting missing values: Usually, you don't want to look at a giant wall of **True/False**. you want to know how many values are missing. you can chain .sum() to end:

```python
df.isna().sum()
```

- the **dropna()** is used to filter out rows or columns that contain NaN or null values. Because most machine learning models and statistical functions can't handle missing data, this is one of the most common steps in data cleaning.

```python
# 2. Drop rows with any missing values
clear_off = df.dropna()
print(clear_off)
```

- If you want to find the data that is actually there, use **notna()** (or its alias **notnull()**)

- While **.dropna()** simply throws away rows with missing values, imputation uses logic or statistics to replace those NaN values with a reasonable guess. This allows you to keep your dataset size intact, which is crucial when your data is limited or when the missingness follows a pattern.

```python
# 3. Imputation: Fill missing values with a logical default
# Here, we fill the missing temperature with the mean of the column
mean_temp = df['Temperature_c'].mean()
df['Temperature_c'] = df['Temperature_c'].fillna(mean_temp)
print(df)
```
