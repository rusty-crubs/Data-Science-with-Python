## 2. The Core Built-in Structures
Python has four primary built-in data structures:

### 1. **List(list)**
- A list is a mutable, ordered sequence of elements.
- In Python, Lists are implemented as **dynamic arrays** (arrays of pointers to objects).
- **When to use**: When you have an ordered collection of homogeneous or heterogeneous items and you need to iterate, append, or modify them frequently.
- **Performance Note**: Appending to the end of a list is fast - an amortized time complexity of O(1). However, inserting or deleting an element at the beginning of a list requires shifting every subsequent element in memory, resulting in O(n) time.

```python
# List Initialization and Operations
sensor_readings = [10.1, 11.3, 12.5, 13.2, 24.5, 25.1, 26.3, 27.8, 28.9, 29.4]

print(f"Sensor Readings: {sensor_readings}")
```

### I. Insertion and Indexing: insert(index, value)
- Indexing is a data structure technique used to speed up the retrival of information.
- **Without an Index**: You would have to walk past every single shelf, reading every book title until you found the one you wanted.
- **With an Index**: You go to the card catalog (the index), fond the book's location code, and walk directly to the correct shelf.

Positive Indexing:

```python
print(f"First Sensor Reading: {sensor_readings[0]}")
print(f"second Sensor Reading: {sensor_readings[1]}")
```

Negative Indexing:

```python
print(f"Last Sensor Reading: {sensor_readings[-1]}")
print(f"Second to Last Sensor Reading: {sensor_readings[-2]}")
```

- Inserting refers to placing a new element into a list at a **specific index**.
- Unlike the **append()** method (which always adds an item to the very end), the **insert()** method allows you to "cut in line" and place an item anywhere you want.

```python
# Inserting at index 0 is inefficient
sensor_readings.insert(0, 9.8)

print(f"Sensor Readings after Insertion: {sensor_readings}")
```

### II. Appending: append()
- Appending is the process of adding a single item to the very end of an existing list.
- The **append()** method takes exactly one argument: the item you want to add.

> list_name.append(item)

```python
# Appending is efficient
sensor_readings.append(30.2)

print(f"Updated Sensor Readings: {sensor_readings}")
```

```python
# List Comprehension: The Pythonic Way to Create Lists
squared_readings = [(x - (sum(sensor_readings)/len(sensor_readings))/x)**2 for x in sensor_readings]
print(f"Squared Sensor Readings: {squared_readings}")
```

### III. Extension
Extend the list by appending all the items from the iterable. Similar to a[len(a):] = iterable.

```python
sensor_readings.extend([1,10])
print(f"Sensor Readings after Extension: {sensor_readings}")
```

### IV. Removing
Remove the first item from the list whose value is the equal to x. It raises a ValueError if there is no such item.

```python
# Removing Elements from a List error
sensor_readings.remove(100)
print(f"Sensor Readings after Removing 1: {sensor_readings}")
```

```python
# Removing Elements from a List
sensor_readings.remove(557.2)
print(f"Sensor Readings after Removing 557.2: {sensor_readings}")
```

### V. POP
- Remove the item at the given position in the list, and return it. 
- If no index is specified, removes and returns the last item in the list.
- It raises *IndexError* if the list is empty or the index is outside the list range.

```python
# Pop removes the last element by default
last_reading = sensor_readings.pop()
print(f"Popped Sensor Reading: {last_reading}")
```

```python
# Pop removing a specific index
reading = sensor_readings.pop(6)
print(f"Popped Sensor Reading: {reading}")
```

### VI. Clear
- Removes all items from the list. Similar to del a[:] (SQL)

```python
# clear method to empty the list
sensor_readings.clear()
print(f"Sensor Readings after Clearing: {sensor_readings}")
```

### VII. Count
- Returns the number of times x appears in the list.

```python
print(f"Count of 10.1 in Sensor Readings: {sensor_readings.count(10.1)}")
```

### VIII. Sorting
- **sort()** Sort the items of the list in place (the arguments can be used for sort customization).

```python
sensor_readings.sort()
print(f"Sorted Sensor Readings: {sensor_readings}")
```

### IX. Reverse
- **reverse()**, reverse the elements of the list in place.

```python
sensor_readings.reverse()
print(f"Reversed Sensor Readings: {sensor_readings}")
```

### X. Copy
- **copy()**, returns a shallow copy of the list. similar to a[:].

```python
copied_readings = sensor_readings.copy()
print(f"Copied Sensor Readings: {copied_readings}")
```
