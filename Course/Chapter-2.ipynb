**elif** statement in Python stands for "else if". It allows us to check multiple conditions, providing a way to execute different blocks of code based on which conditions are true. ## Iterating by Index of Sequences
You can also use the index of elements in the sequence to iterate. The key idea is to first calculate the length of the list and then iterate over the sequence within the range of this length.

```python
li = ['kolar', 'gold', 'field']
for index in range(len(li)):
    print(index, li[index])
```

## While Loop
In Python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes False, the line immediately after the loop in the program is executed.

In below code, loop runs as long as the condition **count < 3** is True. It increments the counter by 1 on each iteration and prints "Hello Students" three times.

```python
count = 0
while count < 3:
    print("Hello Students")
    count += 1
```

## Nested Loops
Python programming language allows to use one loop inside another loop which is called nested loop. For example:

```python
for i in range(1,5):
    for j in range(i):
        print(i, end=" ")
    print()
```

In above code, we used nested loops to print the value of i multiple times in each row, where the number of times it prints **i** increases with each iteration of output loo. The **print()** function prints the value of **i** and moves to the next line after each row.

A final note on loop nesting is that we can put any type of loop inside of anu other type of loops in Python. For example, a for loop can inside a while loop or vice versa.

## Loop Control Statement
Loop control statements change execution from their normal sequence. When execution leaves a scope, all automatic objects that were created in that scope are destroyed. Python supports the following control statement.

## Continue Statement
The continue statement in Python returns the control to the beginning of the loop.

```python
for letter in "Kolar Gold Field":
    if letter == 'o':
        continue
    print(letter, end=" ")
```

The continuous statement is used to skip the current iteration of a loop and move to next iteration. It is useful when we want to bypass certain conditions without terminating the loop.

## Break Statement
Break statement is used to exit the loop prematurely when a specified condition is met. The break statement in Python brings control out of the loop when certain conditions are met.

```python
for letter in "Kolar Gold Field":
    if letter == 'o':
        break
    print(letter, end=" ")
```

In this code, we are using a for loop to iterate through each letter in the string "Kolar Gold Field". The loop checks if the current letter is 'o'. If it is, the loop will skip the rest of the code in that iteration and move on to the next letter (using continue). If the letter is not 'o', it will print the letter followed by a space (using end=" " to avoid moving to a new line).

## Pass Statement
The pass statement is a null operation; it does nothing when executed. It is used as a placeholder in situations where syntactically some code is required, but you do not want to execute any code. For example, you can use it in an empty function or an empty loop.

```python
for letter in "Kolar Gold Field":
    pass
print('Last statement executed.:', letter)
```

In this example, we are using the pass statement as a placeholder in the for loop. The loop iterates through each letter in the string "Kolar Gold Field", but since there is no code inside the loop (only the pass statement), it does nothing for each iteration. After the loop completes, it prints the last letter that was iterated over, which is 'd'.

Functions are reusable blocks of code that perform a specific task. They help in breaking down complex problems into smaller, manageable pieces and promote code reusability.

Defining a function allows you to encapsulate a segement of code, specifying the information that enters and leaces the code. You can make use of this "code-capsule" repeatedly and in many different contexts. For example, suppose you want to count how many vowels are in a string. The following defines a function that accomplishes this:

```python
def count_vowels(in_string):
    num_vowels = 0
    vowels = 'aeiouAEIOU'
    for char in in_string:
        if char in vowels:
            num_vowels += 1
    return num_vowels
```

Executing this code will define the function named **count_vowels**. This function expects to be passed one object, repeated by **in_string**, as an input argument, and it will return number of vowels stored in that object. Invoking **count_vowels**, passing it an input object, is referred to as calling the function:

```python
count_vowels("Kolar Gold Field")
```

The great thing about it is that it can be used over and over again!

```python
count_vowels("Data Science with Python")
```

```python
count_vowels("Billie jean not my lover")
```

## The **def** statement
similar to **if**, **else** and **for**, the **def** statement is reserved by the Python Language to signify the definition of functions (and a few other things that we'll be covering later). The following is the general syntax for defining a Python function: 

 def function_name (function_significance):

  """"" Documentation String """"""

  encapulated_code

  return object

**function_name**: can be anu valid variable name, and must be followed by parentheses and then a colon.

**function_significance**: specifies the input arguments to the function, and may be left blank if the function does not accept any argument (the parentheses must still be included, but will not encapsulate anything).

**Encapsulated_code**: can consists of general Python code, and is demarcated by being indented relative to the **def** statement.

**return**: if reached by the encapsulated code, triggers the function to return the specified object and end its own executation immediately.

The **return** statement is also reserved by Python. It denotes the end of a function; if reached, a return statement immediately concludes the execution of the function and returns the specified object.

## The **return** Statement
In general, any Python object can follow a function's **return** statement. Furthermore, an empty **return** statement can be specified, or the return statement of a function can be omitted together. In both of these cases, the function will return the **None** object.

```python
# this function does nothing but is syntactically correct
# This function returns None by default if there is no return statement
def f():
    x = 1
    return
```

```python
# return statement is omitted, so it returns None by default
def f():
    x = 1
```

```python
def count_vowels(in_string):
    # Returns the number of vowels in the in_string
    return sum(1 for char in in_string if char in 'aeiouAEIOU')
```

## Multiple **return** statements
you can specify more than one **return** statement within a function. This can be useful for handling edge-cases or ooptimizations in your code. Suppose you want your function to compute $e^x$, using a **Taylor series** approximation. The function should immediately return *1.0* in the case that x=0:

```python
from math import factorial
def compute_exp(x):
    if x == 0:
        return 1.0
    else:
        return sum(x**n/factorial(n) for n in range(1, 100))
```

If **x==0** is *True*, then the first **return** statement is reached. *1.0* will be returned and the function will be "exited" immediately, without ever reaching the code following it.

As stated above, a **return** statement will trigger a function to end its execution immediately when reached, even subsequent code follow it. It is impossible for multiple 'return' statements to be visited within a single function must return a single container of those items, like a list or a tuple.

```python
# Returning multiple items from a function
def bad_func(x):
    # return x**2, x**3
    return x**2
    return x**3
```

```python
def good_func(x):
    return x**2, x**3
```

```python
bad_func(2)
```

```python
good_func(2)
```

## Arguments
A sequence of comma-seperated variable names can be specified in the function signature to indicated positionl arguments for the function. For example, the following specifies *x*, *lower*, and *upper* as input arguments to function, **is_bounded**:

```python
def is_bounded(x,lower,upper):
    return lower <= x <= upper
```

This function can then be passed its arguments in several ways:

### Specifying Arguments by Position
The objects passed to **is_bounded** will be assigned to its input variables based on their positions. That is, **is_bounded(3,2,4)** will assign **x=3**, **lower=2**, and **upper=4**, in accordance with the positional ordering of the function's input arguments:

```python
# evaluate: 2 <= 3 <= 4
# Specifying inputs based on position
is_bounded(3,2,4)
```

Feeding a function too few or too many arguments will raise a **TypeError**

```python
# too few input: raise error
is_bounded()
```

```python
# too many inputs: raise error
is_bounded(3,2,4,5)
```

### Specifying Arguments by Name
You can provide explicit names when specifying the inputs to a function, in which case ordering does not matter. This is very nice for writing clear and flexible code:

```python
# evaluate: 2 <= 3 <= 4`
# specifying inputs based on name
is_bounded(x=3, lower=2, upper=4)
```

You can mix-and-match positional and named input by using position-based inputs first:

```python
# evaluate: 2 <= 3 <= 4
# 'x' is specified based on position, 'lower' and 'upper' are specified based on name
is_bounded(3, lower=2, upper=4)
```

📝Note: if you provide a named input, all inputs following it must also be named:

```python
# Position arguments cannot follow named arguments
is_bounded(lower=2, upper=4, 3)
```

### Default-Valued Arguments
You can specify default values for input arguments to a function. Their default values are utilized if a user does not specify these inputs when calling the function. Recall our **count_vowels** function. Suppose we want the ability to include 'y' as a vowel. We know, however, that people will typically want to exclude 'y' from their vowels, so we can exclude 'y' by default:

```python
def count_vowels(in_string, include_y=False):
    vowels='aeiouAEIOU'
    if include_y:
        vowels += 'yY'
    return sum(1 for char in in_string if char in vowels)
```

Now, if only **in_string** is specified when calling **count_vowels**, **include_y** will be passed the value **False** by default:

```python
# using the default value: exclude y from vowels
count_vowels("Happy")
```

This default value can be overridden:

```python
# overriding the delault value: include y in vowels
count_vowels("Happy", include_y=True)
```

```python
# you can still specifu inputs by name
count_vowels(in_string="Happy", include_y=True)
```

Default-valued input arguments must come after all positional input arguments in the function signature:

```python
# this is ok
def f(x,y,z,count=1,upper=2):
    return None
```

```python
# this will raise a SyntaxError: positional argument follows keyword argument
def f(x,y,count=1,upper=2,z):
    return None
```

### Accomodating an Arbitrary Number of Positional Arguments
Python provides us with a syntax for defining a function, which can be called with an arbitrary number of positional arguments. This is signaled by the syntax **def f(*<var_name>)**.

```python
# The * symbol indicates that an arbitrary number of arguments can be passed to `args`, when calling `f`.
def f(*args):
    # All arguments passed to `f` will be collected into a tuple that is assigned to the variable `args`
    # `f()` will assign `args = tuple()`
    # `f(1,2,3)` will assign `args = (1,2,3)`
    return args
```

Because Python cannot foresee how many arguments will be passed to **f**, all of the objects that are passed to it will be packed into a tuple, which is then assigned to the variable **args**:

```python
# pass zero arguments to `f`
f()
```

```python
# pass one argument to `f`
f(1)
```

```python
# Pass three arguments to `f`
f((0,1), True, "Cow")
```

This syntax can be combined with positional arguments and default arguments. Any variables specified after a packed variable must be called by a name:

```python
def f(x,*seq,y):
    print(f"x is {x}")
    print(f"seq is {seq}")
    print(f"y is {y}")
    return None
```

```python
f(1, 2, 3, 4, y=5) # `y` must be specified by name
```

```python
f("cat",y="dog")
```

```python
def mean(*seq):
    total = sum(seq)
    count = len(seq)
    return total / count if count > 0 else 0
```

```python
mean(1,2,3)
```

we see that * indicates the packing of an arbitrary number of arguments into a tuple, when used in the signature of a function definition. Simultaneously, * signals the unpacking of an iterable to pass each of ots members as a positional argument to a function, when used in context of calling a function:

```python
def f(x,y,z): return x+y+z
```

```python
f(1,2,3)
```

```python
# * means that all remaining positional arguments will be collected into a tuple and assigned to `seq`
f(*[1,2,3])
```

```python
def number_of_args(*args):
    return len(args)
```

```python
number_of_args(None,None,None,None)
```

```python
some_list = [1, 2, 3, 4, 5]
```

```python
# Passing the list itself as the sole argument
number_of_args(some_list)
```

```python
# unpacking the 5 members of the list,
# passing each one as an argument to the function
number_of_args(*some_list)
```

### Accommodating an Arbitrary Number of Keyword Arguments
We can also define a function that is able to accept an arbitrary number of keyword arguments, using the syntax:
> def f(**<var_name>)

📝Note: A single asterisk *, was used to denote an arbitrary number of positional arguments, whereas double asterisk ** signals the accpetance of an arbitrary number of keyword arguments.

```python
# The ** symbol indicates that an arbitrary number of keyword arguments can be passed to `kwargs`, when calling `f`.
def f(**kwargs):
    # All keyword arguments passed to `f` will be collected into a dictionary that is assigned to the variable `kwargs`
    # `f()` will assign `kwargs = dict()`
    # `f(x=1, y=2)` will assign `kwargs = {'x': 1, 'y': 2}`
    return kwargs
```

Because Python cannot forsee how many arguments will be passed to **f**, all of the keyword arguments that are passed to it will be packed into a dictionary, where a given keyword is set as a key (cast as a string) that maps to the corresponding value. This dictionary is then assigned to the variable **kwargs**.

```python
f() # pass zero keyword arguments to `f`
```

```python
f(x=1) # pass one keyword argument to `f`
```

```python
f(name='Bhim Rai Thapa')
```

```python
f(x=(0,1), val=True, moo="cow") # pass three arguments to `f`, each specified by name
```

This syntax can be combined with positional arguments and default arguments. No additional arguments may come after a ** entry in a function-definition signature:

```python
def f(x,y=2,**kwargs):
    print(f"x is {x}")
    print(f"y is {y}")
    print(f"kwargs is {kwargs}")
    return None
```

```python
# Passing arbitrary number of keyword arguments to `f`
f(1, y=99, val=True, moo="cow") 
```

```python
# no additional keyword arguments passed to `f`
f("cat", y="dog")
```

The following function accepts an arbitrary number of positional arguments and an arbitrary number of keyword argument:

```python
# accepting arbitrary positional and keyword arguments
def f(*args, **kwargs):
    # all positional arguments passed to `f` will be collected into a tuple that is assigned to `args`
    # all keyword arguments passed to `f` will be collected into a dictionary that is assigned to `kwargs`
    print(f"args is {args}")
    print(f"kwargs is {kwargs}")
    return None
```

```python
f(1, 2, 3, hi=-1, bye=-2, sigh=-3)
```

We see that ** indicates the packing of an arbitrary number of keyword arguments into a dictionary, when used in the signature of a function definition. Simultaneously, ** signals the unpacking of a dictionary to pass each of its key-value pairs as keyword argument to a function, when used in the context of calling a funtion:

```python
# using ** when calling a function allows you to unpack a dictionary and pass its key-value pairs as keyword arguments to the function
def f(x,y,z):
    return 0*x + 1*y + 2*z

# Example usage:
params = {"x": 9, "y": 1, "z": 10}
result = f(**params)
```

```python
result
```

- call the function, passing it an arbitrary number of keyword arguments, by unpacking a dictionary.

```python
def print_kwargs(**kwargs):
    print(f"kwargs is {kwargs}")
```
