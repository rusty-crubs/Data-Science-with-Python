Conditional programming, or branching, is something we do in our day-to-day task or evry moment. It's about evaluating conditions: if the path light is green, then you can cross the road, if it's raining, then I'm taking umbrella, and so on.
Conditional statement in Python are used to execute certain blocks of code based on specific conditions. These statements help control the flow of a program, making it behave differently in different situations.


## If Conditional Statement
**Learning Objective**

By the end of the section you will be able to:
- Understand the concepts of if statement.
- how to use it.


The main tool is the **if** statement, which comes in different forms and colors, but basically what it does is evaluate an expression and, based on the result, choose which part of the code to execute. Example:

```python
late = True
if late:
    print("I need to call my manager to reschedule my meeting.")
```

This is possibly the simplest example: when fed to the if statement, *late* acts as a conditional expression, which is evaluated in a Boolean context (exactly like if we were calling bool (*late*)). If the result of the evaluation is **True**, then we enter the body of code immediately after the if statement. Notice that the **print** instruction is indented: this means it belongs to a scope defined by the if clause.


## If else Conditional Statement
**Learning Objective**

By the end of this section, you should be able to:
- Understand the concept and use of if-else statement.
- Also shorthand methods for if-else statement.

**if else** allows us to specify a block of code that will execute if the condition(s) associated with an if or elif statement evaluates to False. Else block provides a way to handle all other cases that don't meet the specified conditions.

```python
age = 10
if age < 18:
    print("You are a minor.")
else:    print("You are an adult.")
```

### Short hand if-else statement
The short-hand if-else statement allows us to write a single-line if-else statement.

```python
mark = 45
res = "Pass" if mark >= 40 else "Fail"
print(res)
```

📝Note: This method is also known as ternary operator. Ternary Operator essentially a shorthand for the if-else statement that allows us to write more compact and readable code, especially for simple conditions.

## elif Statement
**Learning Objective**

By the end of this section, you should be able to:
- Understand the concept and use of if-elif-else statement.

**elif** statement in Python stands for "else if". It allows us to check multiple conditions, providing a way to execute different blocks of code based on which conditions are true. Using elif statement makes oure code more readable and efficient by eliminating the need for multiple nested if-else statement or if statements.

```python
age = 35
if age <= 12:
    print("You are a child.")
elif age <= 18:
    print("You are a teenager.")
elif age <= 35:
    print("You are a young adult.")
else:
    print("You are an adult.")
```

In this example, the code checks the value of age using if-elif-else. Since age is 36, it skips first two conditions (age<=12 and age<=18), and the third condition (age<=35) is True, so it prints "You are a young adult".

## Nested if-else Statement
**Learning Objective**

By the end of this section, you should be able to:
- understand the concepts and use of nested if-else statement.

Nested if-else statements means if-else statement inside another if-else statement. We can use nested if statements to check conditions within conditions.

```python
age = 70
is_member = True

if age >= 65:
    if is_member:
        print("30% senior discount applied.")
    else:
        print("20% senior discount applied.")
else:
    print("Not eligible for senior discounts.")
```

In the example, the code checks tha **age** and **is_member** using nested if-else statement. When the condition meets it provides appropriate responses according to the programmed conditions and statements.

## Match-Case Statement
**Learning Object**

By the end of this section, you should be able to:
- understand the cases implications and
- how to impliment it

Match-case Statement in Python is the version of a switch-case found in other languages (like C/C++, Java). It allows us to match a variable's value against a set of patterns.

```python
number = 2

match(number):
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Other number")
```

## The **for** loop
The for loop is used when looping a sequence, like a list, tuple, or a collection of objects. It allow to execute a block of code repeatedlym once for each item in the sequence.

```python
n = 4
for i in range(0,n):
    print(i)
```
